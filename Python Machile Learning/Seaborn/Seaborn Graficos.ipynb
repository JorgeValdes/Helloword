# Seaborn 

seaborn es una libreria de python basada en matplotlib. libreria de visualizacion de alto nivel para realizar graficos estadisticos de forma vistosa.

```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
propinas = sns.load_dataset('tips')
```

```python
propinas.describe()
```

```python
sns.distplot(propinas['total_bill'], bins=20)
```

```python
sns.distplot(propinas['tip'], bins=20)
```

```python
sns.distplot(propinas['size'], bins=20)
plt.title('grafico size')
```

```python
#propinas
sns.jointplot(x='total_bill', y='tip', data=propinas)
```

```python
sns.jointplot(x='total_bill', y='tip', data=propinas, kind='hex')
```

```python
sns.jointplot(x='total_bill', y='tip', data=propinas, kind='reg')
```

```python
sns.jointplot(x='total_bill', y='tip', data=propinas, kind='kde')
```

```python
sns.pairplot(propinas)
```

```python
sns.pairplot(propinas)
```

```python
sns.pairplot(propinas, hue='sex')
```

```python
sns.rugplot(propinas['total_bill'])
```

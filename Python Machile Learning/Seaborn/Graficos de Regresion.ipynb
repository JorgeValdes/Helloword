Graficos de regresion(Regression Plots)

```python
import seaborn as sns
```

```python
propinas = sns.load_dataset('tips')
```

```python
propinas.head()
```

```python
sns.lmplot(x='total_bill', y='tip', data=propinas, hue='sex') 
#grafico para regresion lineal
```

```python
sns.lmplot(x='total_bill', y='tip', data=propinas, hue='sex', markers=['o', 'x'], scatter_kws = {'s':180}) 
```

```python
sns.lmplot(x='total_bill', y='tip', data=propinas, col='sex')
```

```python
sns.lmplot(x='total_bill', y='tip', data=propinas, aspect=0.8, size=10)
```

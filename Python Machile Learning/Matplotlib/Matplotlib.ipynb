```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(2,9,10)
```

```python
x
```

```python
y = x**2
```

```python
y 
```

```python
plt.plot(x,y, 'red')
plt.title("Este es mi grafico")
plt.ylabel("eje y")
plt.xlabel("eje x")
```

```python
plt.subplot(2,1,1)
plt.plot(x,y, 'green')

plt.subplot(2,1,2)
plt.plot(y,x,'red')
```

```python
#Programacion orientada objetos
figura = plt.figure()
grafico = figura.add_axes([ 0.1,0.1,0.8,0.8])
grafico.plot(x,y, 'green')
```

```python
figura = plt.figure()
grafico1 = figura.add_axes([0.1,0.1,0.9,0.9])
grafico1.plot(x,y,'red')

grafico2 = figura.add_axes([0.2,0.6,0.3,0.3])
grafico2.plot(y,x , 'blue')
```

```python
#Multigraficos
figura, graficos = plt.subplots(nrows =2, ncols =2)
graficos[0][0].plot(x,y,'blue')
graficos[0][1].plot(x,y,'g')
graficos[1][0].plot(x,y,'pink')
graficos[1][1].plot(x,y,'red')

graficos[0][0].set_title('graficos')
```

```python
#Tamaño del Grafico
```

```python
figura = plt.figure(figsize=(8,4))
grafico = figura.add_axes([0.1, 0.1, 0.9, 0.9])
grafico.plot(x,y,'r')
```

##  Dos graficos en la misma figura

```python
figura = plt.figure()
graficos = figura.add_axes([0.1, 0.1, 0.9, 0.9])
graficos.plot(x , x**2, label='Al cuadrado')
graficos.plot(x, x**3, label='al Cubo')
graficos.legend(loc=0)
```

# Color del grafico

```python
figura = plt.figure()
grafico = figura.add_axes([0,0,1,1])
grafico.plot(x,y,color = 'tomato')
```

# Linea del grafico (grosor , transparencia y tipo)

```python
figura = plt.figure()
grafico  = figura.add_axes([0,0,1,1])
grafico.plot(x,y,color = 'red', linewidth=10, alpha=0.5, linestyle='-')
#grafico.title('grafico>')
plt.title('este es mi grafico')
```
